```python
def convert_price(value):
    if 'Cr' in value:
        value = value.replace(' Cr', '')
        value = float(value) * 10**7
    elif 'Lacs' in value:
        value = value.replace(' Lacs', '')
        value = float(value) * 10**5
    elif 'L' in value:
        value = value.replace(' L', '')
        value = float(value) * 10**5
    elif 'K' in value:
        value = value.replace('K', '')
        value = float(value) * 10**3
    
    return value
```

```python
df1['Flat_Price'] = df1['Flat_Price'].apply(convert_price)
```

```python
df1
```

```python
df1['EMI_Starts'] = df1['EMI_Starts'].apply(convert_price)
```

```python
df1['Price_per_sq.ft'] = df1['Price_per_sq.ft'].str.replace('/sq.ft', '')
```

```python
df1['Price_per_sq.ft'] = df1['Price_per_sq.ft'].apply(convert_price)
```

```python
df1.Location=df1.Location.apply(lambda x: x.strip())
location_stats=df1.groupby('Location')['Location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10=location_stats[location_stats<=10]
df1.Location=df1.Location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df1.Location.unique())
```

```python
column_mapping = {
    'Total_Sq.ft': 'Total_Sqft',
    'Price_per_sq.ft': 'Price_per_sqft'
}
df1 = df1.rename(columns=column_mapping)
df1
```

```python
df2=df1[~(df1.Total_Sqft/df1.BHK<300)]
df2.shape
```

```python
df2
```

```python
dummy=pd.get_dummies(df2.Location)
dummy
```

```python
df3=pd.concat([df2,dummy.drop('other',axis='columns')],axis='columns')
df3
```

```python
df3=df3.drop('Location',axis='columns')
df3
```

```python
df3.shape
```

```python
X=df3.drop('Flat_Price',axis='columns')
X=X.drop('EMI_Starts',axis='columns')
X
```

```python
y=df3.Flat_Price
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf=LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
def predict_price(Location,Total_Sqft,BHK):
    loc_index=np.where(X.columns==Location)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=Total_Sqft
    x[1]=BHK
    if loc_index>=0:
        x[loc_index]=1
    return lr_clf.predict([x])[0]
```
